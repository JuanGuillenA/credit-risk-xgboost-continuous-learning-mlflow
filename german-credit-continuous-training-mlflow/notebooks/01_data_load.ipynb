Title: Cooperative Credit Risk MVP — Data Ingestion
Context: A newly founded cooperative starts with limited historical credit applications. We use the German Credit dataset as a public proxy for initial member applications.

```python
import os
import pandas as pd
from sklearn.datasets import fetch_openml

os.makedirs("../artifacts/dataset", exist_ok=True)

data = fetch_openml(name="credit-g", version=1, as_frame=True)
df = data.frame.copy()

print("Dataset shape:", df.shape)
display(df.head())

raw_path = "../artifacts/dataset/german_credit_raw.csv"
df.to_csv(raw_path, index=False)
print("Saved:", raw_path)
```
